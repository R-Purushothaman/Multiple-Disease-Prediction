# Liver

```python
liver=pd.read_csv('/content/indian_liver_patient - indian_liver_patient.csv')
liver.columns = liver.columns.str.lower()
display(liver)
```

```python
liver['albumin_and_globulin_ratio']=liver['albumin_and_globulin_ratio'].fillna(liver['albumin_and_globulin_ratio'].mean())
```

```python
x=liver.drop(['dataset'],axis=1)
y=liver['dataset']

le=LabelEncoder()
x['gender']=le.fit_transform(x['gender'])
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from imblearn.under_sampling import RandomUnderSampler
rus = RandomUnderSampler()

x_train_rus, y_train_rus = rus.fit_resample(x_train,y_train)
```

```python
rf=RandomForestClassifier(n_estimators=100)
rf.fit(x_train_rus,y_train_rus)
```

```python
feature_importances = pd.Series(rf.feature_importances_, index=x.columns).sort_values(ascending=False)
print("Feature Importances:\n",feature_importances)
```

```python
x=liver[['alkaline_phosphotase','aspartate_aminotransferase','alamine_aminotransferase','age','direct_bilirubin','total_bilirubin']]
y=liver['dataset']

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
rt=RandomUnderSampler()
x_train_rt,y_train_rt=rt.fit_resample(x_train,y_train)
```

```python
final=RandomForestClassifier(n_estimators=250)
final.fit(x_train_rt,y_train_rt)
```

```python
yp=final.predict(x_test)
print(classification_report(y_test,yp))
```

```python
# pickling the file
import pickle
pickle.dump(final,open('liver.pkl','wb'))
```

# parkinsons

```python
parkinsons=pd.read_csv('/content/parkinsons - parkinsons.csv')
display(parkinsons)
```

```python
parkinsons['status'].value_counts()
```

```python
le=LabelEncoder()
parkinsons['name']=le.fit_transform(parkinsons['name'])
```

```python
x=parkinsons.drop(['status'],axis=1)
y=parkinsons['status']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from imblearn.under_sampling import RandomUnderSampler
us = RandomUnderSampler(random_state=42)
x_train_us, y_train_us = us.fit_resample(x_train,y_train)
```

```python
df=RandomForestClassifier(n_estimators=100)
df.fit(x_train_us,y_train_us)
```

```python
feature_importances= pd.Series(df.feature_importances_, index=x_train.columns).sort_values(ascending=False)
print("Feature Importances:\n",feature_importances)
```

```python
# selecting top 6 features
x=parkinsons[['PPE','spread1','MDVP:APQ','Jitter:DDP','MDVP:Fhi(Hz)','MDVP:Fo(Hz)']]
y=parkinsons['status']

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
rs=RandomUnderSampler()
x_train_rs,y_train_rs=rs.fit_resample(x_train,y_train)
```

```python
model=RandomForestClassifier(n_estimators=300)
model.fit(x_train_rs,y_train_rs)
```

```python
y_pred=model.predict(x_test)
print(classification_report(y_test,y_pred))
```

```python
import pickle
pickle.dump(model,open('parkinsons.pkl','wb'))
```

# Kidney

```python
kidney=pd.read_csv('/content/kidney_disease - kidney_disease.csv')
kidney.head(10)
```

```python
kidney['pcv'] = pd.to_numeric(kidney['pcv'], errors="coerce").astype(float)
kidney['wc'] = pd.to_numeric(kidney['wc'], errors="coerce").astype(float)
kidney['rc'] = pd.to_numeric(kidney['rc'], errors="coerce").astype(float)
```

```python
kidney["pcc"] = kidney["pcc"].fillna(kidney["pcc"].mode()[0])
kidney["ba"] = kidney["ba"].fillna(kidney["ba"].mode()[0])
```

```python
kidney.info()
```

```python
kidney.isnull().sum()
```

```python
num_cols = kidney.select_dtypes(include=['int64','float64']).columns
cat_cols = kidney.select_dtypes(include=['object']).columns

print("Numerical Columns:\n", num_cols)
print("\nCategorical Columns:\n", cat_cols)
```

```python
for col in ['age', 'sg', 'hemo', 'pcv', 'rc','bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'wc']:
    kidney[col] = kidney[col].fillna(kidney[col].median())

for col in ['rbc', 'pc', 'htn', 'dm', 'cad', 'appet', 'pe', 'ane']:
    kidney[col] = kidney[col].fillna(kidney[col].mode()[0])

print(kidney.isnull().sum())
```

```python
import numpy as np

kidney.replace(['?', ' ?', '\t?', 'nan', 'NaN', ''], np.nan, inplace=True)
```

```python
le=LabelEncoder()
kidney['rbc']=le.fit_transform(kidney['rbc'])
kidney['pc']=le.fit_transform(kidney['pc'])
kidney['htn']=le.fit_transform(kidney['htn'])
kidney['dm']=le.fit_transform(kidney['dm'])
kidney['cad']=le.fit_transform(kidney['cad'])
kidney['appet']=le.fit_transform(kidney['appet'])
kidney['pe']=le.fit_transform(kidney['pe'])
kidney['ane']=le.fit_transform(kidney['ane'])
kidney['pcc']=le.fit_transform(kidney['pcc'])
kidney['ba']=le.fit_transform(kidney['ba'])
kidney['classification']=le.fit_transform(kidney['classification'])
```

```python
x=kidney.drop(['classification'],axis=1)
y=kidney['classification']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
kidney['classification'].value_counts()
```

```python
from imblearn.under_sampling import RandomUnderSampler
rus = RandomUnderSampler()

x_train_rus, y_train_rus = rus.fit_resample(x_train,y_train)
```

```python
ra=RandomForestClassifier(n_estimators=100)
ra.fit(x_train_rus,y_train_rus)
```

```python
feature_importances = pd.Series(ra.feature_importances_, index=x.columns).sort_values(ascending=False)
print("Feature Importances:\n",feature_importances)
```

```python
# selecting the features

x=kidney[['hemo','pcv','age','sg','sc','rbc','pc','wc','bgr']]
y=kidney['classification']


x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=5)
```

```python
final=RandomForestClassifier(n_estimators=10)
final.fit(x_train,y_train)
```

```python
yp=final.predict(x_test)
print(classification_report(y_test,yp))
```

```python
# pickling the model

import pickle
pickle.dump(final,open('kidney.pkl','wb'))
```
